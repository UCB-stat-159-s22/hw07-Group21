```python
def plot_gender_for_major_by_age(fod1p_code):
    majors_df = all_US.query('FOD1P == @fod1p_code | FOD2P == @fod1p_code')
    bins= [5*n for n in range(3,21)]
    majors_df['AgeGroup'] = pd.cut(majors_df['AGEP'], bins=bins, right=False)
    counts = majors_df.groupby(['AgeGroup','SEX']).count()['Major'].reset_index()
    sns.catplot(x = "AgeGroup",       # x variable name
                y = "Major",       # y variable name
                hue = "SEX",  # group variable name
                data = counts,     # dataframe to plot
                kind = "bar");
```

```python
def plot_gender_ratio_for_major_by_age(fod1p_code):
    majors_df = all_US.query('FOD1P == @fod1p_code | FOD2P == @fod1p_code')
    bins= [5*n for n in range(4,20)]
    majors_df['AgeGroup'] = pd.cut(majors_df['AGEP'], bins=bins, right=False)
    counts = majors_df.groupby(['AgeGroup','SEX']).count()['Major'].reset_index()
    totals_by_age = counts.groupby('AgeGroup').sum()['Major'].reset_index().rename(columns={'Major':'Total'})
    counts = counts.merge(totals_by_age, on='AgeGroup')
    counts['Percent'] = counts['Major']/counts['Total']
    counts.query('SEX == 2').plot(x='AgeGroup',y='Percent')
```

```python
plot_gender_ratio_for_major_by_age(2102)
```
